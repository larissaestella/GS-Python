```python
from tkinter import *
from tkinter import Tk, ttk
from tkinter import messagebox


# cores -----------------------------
co0 = "#f0f3f5"  # Preta 
co1 = "#feffff"  # Branca 
co2 = "#3fb5a3"  # Azul
co3 = "#38576b"  # Valor 
co4 = "#403d3d"   # Letra 


# criando janela ----------------------------
janela = Tk()
janela.title('Acesso ao Cadastro do Paciente')
# janela.geometry('310x300')
janela.configure(background=co1)
# janela.resizable(width=FALSE, height=FALSE)

# dividindo a  janela ----------------------------
frame_cima = Frame(janela, width=610, height=50, bg=co1, relief='flat')
frame_cima.grid(row=0, column=0, pady=1, padx=0, sticky=NSEW)

frame_baixo = Frame(janela, width=610, height=450, bg=co1, relief='flat')
frame_baixo.grid(row=1, column=0, pady=1, padx=0, sticky=NSEW)

# configurando frame cima ----------------------------
l_nome = Label(frame_cima, text='Acesso', anchor=NE, font=('Ivy 25'), bg=co1, fg=co4)
l_nome.place(x=15, y=5)

l_linha = Label(frame_cima, text='', width=580,  anchor=NW, font=('Ivy 1'), bg=co2, fg=co4)
l_linha.place(x=10, y=45)

# criando funções ----------------------------

## função para  verificar senha
credenicais = ['Admin', '123456']

def verificar_senha():
    nome = e_nome.get()
    senha = e_senha.get()

    if credenicais[0] == nome and credenicais[1] == senha:
        messagebox.showinfo('Login', 'Seja Bem Vindo de Volta ' +credenicais[0])
       
        # deletar itens presentes no frame baixo e cima 
        for widget in frame_baixo.winfo_children():
            widget.destroy()
        for widget in frame_cima.winfo_children():
            widget.destroy() 
        nova_janela()

    else:
        messagebox.showwarning('Erro', 'Verifique o login e senha')

## função após verificação 
def nova_janela():
    # configurando frame cima 2 ----------------------------
    l_nome = Label(frame_cima, text='Usuário: ' +credenicais[0], anchor=NE, font=('Ivy 20'), bg=co1, fg=co4)
    l_nome.place(x=15, y=5)

    l_linha = Label(frame_cima, text='', width=580,  anchor=NW, font=('Ivy 1'), bg=co2, fg=co4)
    l_linha.place(x=10, y=45)


    ## função para acessar cadastro do paciente 
        # Lista de pacientes
    lista_pacientes = []

    def acessar_paciente():
        nome = e_nome.get()
        cpf = e_cpf.get()
        data = e_data.get()
        horario = e_horario.get()

        paciente = {f'Nome': nome, 'CPF': cpf,'Data da Consulta': data, 'Horário': horario}
        lista_pacientes.append(paciente)


        for widget in frame_baixo.winfo_children():
            widget.destroy()
        nova_janela2()

    ## função após verificação 
    def nova_janela2():
        nome = e_nome
             # configurando frame baixo 3
        l_nome = Label(frame_baixo, width=100, text='Detalhes do Paciente:\n ' + str(lista_pacientes), anchor=NW, font=('Ivy 10'), bg=co1, fg=co4)
        l_nome.place(x=10, y=20)

    # configurando frame baixo 2----------------------------

    l_nome = Label(frame_baixo, text='NOME E SOBRENOME:', anchor=NW, font=('Ivy 10'), bg=co1, fg=co4)
    l_nome.place(x=10, y=20)
    e_nome = Entry(frame_baixo, width=25, justify='left', font=('', 15), highlightthickness=1,relief='solid')
    e_nome.place(x=14, y=50)

    l_cpf = Label(frame_baixo, text='CPF:', anchor=NW, font=('Ivy 10'), bg=co1, fg=co4)
    l_cpf.place(x=10, y=95)
    e_cpf = Entry(frame_baixo, width=25, justify='left', font=('', 15), highlightthickness=1,relief='solid')
    e_cpf.place(x=14, y=130)

    l_data = Label(frame_baixo, text='DATA DA CONSULTA:', anchor=NW, font=('Ivy 10'), bg=co1, fg=co4)
    l_data.place(x=10, y=180)
    e_data = Entry(frame_baixo, width=25, justify='left', font=('', 15), highlightthickness=1,relief='solid')
    e_data.place(x=14, y=210)

    l_horario = Label(frame_baixo, text='HORÁRIO:', anchor=NW, font=('Ivy 10'), bg=co1, fg=co4)
    l_horario.place(x=10, y=265)
    e_horario = Entry(frame_baixo, width=25, justify='left', font=('', 15), highlightthickness=1,relief='solid')
    e_horario.place(x=14, y=290)

    b_botao = Button(frame_baixo, command=acessar_paciente,text='Consultar ', width=39, height=2, font=('Ivy 8 bold'), bg=co2, fg=co1, relief=RAISED, overrelief=RIDGE)
    b_botao.place(x=15, y=340)

# configurando frame baixo ----------------------------

l_nome = Label(frame_baixo, text='Login *', anchor=NW, font=('Ivy 10'), bg=co1, fg=co4)
l_nome.place(x=10, y=20)
e_nome = Entry(frame_baixo, width=25, justify='left', font=('', 15), highlightthickness=1,relief='solid')
e_nome.place(x=14, y=50)

l_senha = Label(frame_baixo, text='Senha *', anchor=NW, font=('Ivy 10'), bg=co1, fg=co4)
l_senha.place(x=10, y=95)
e_senha = Entry(frame_baixo, width=25, justify='left', show='*', font=('', 15), highlightthickness=1,relief='solid')
e_senha.place(x=14, y=130)

b_botao = Button(frame_baixo, command=verificar_senha,text='Entrar', width=39, height=2, font=('Ivy 8 bold'), bg=co2, fg=co1, relief=RAISED, overrelief=RIDGE)
b_botao.place(x=15, y=180)

janela.mainloop()
```
